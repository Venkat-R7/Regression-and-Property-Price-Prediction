```python
## Defining one hot encoder
encoder = OneHotEncoder(handle_unknown='ignore')

## Encoding County column
encoder_df = pd.DataFrame(encoder.fit_transform(ppr_data[['County']]).toarray())
encoder_df.columns = encoder.get_feature_names_out(['County'])
final_df = ppr_data.join(encoder_df)

## Encoding Not Full Market Price column
encoder_df = pd.DataFrame(encoder.fit_transform(final_df[['Not Full Market Price']]).toarray())
encoder_df.columns = encoder.get_feature_names_out(['Not Full Market Price'])
final_df = final_df.join(encoder_df)

## Encoding VAT Exclusive column
encoder_df = pd.DataFrame(encoder.fit_transform(final_df[['VAT Exclusive']]).toarray())
encoder_df.columns = encoder.get_feature_names_out(['VAT Exclusive'])
final_df = final_df.join(encoder_df)

## Encoding Property description column
encoder_df = pd.DataFrame(encoder.fit_transform(final_df[['Description of Property']]).toarray())
encoder_df.columns = encoder.get_feature_names_out(['Description of Property'])
ppr_data_f = final_df.join(encoder_df)

## Dropping original columns from the DF
ppr_data_f.drop('County', axis=1, inplace=True)
ppr_data_f.drop('Not Full Market Price', axis=1, inplace=True)
ppr_data_f.drop('VAT Exclusive', axis=1, inplace=True)
ppr_data_f.drop('Description of Property', axis=1, inplace=True)
```

#### Feature extraction on datetime column

```python
## Converting the date of sale column datatime datatype from string
ppr_data_f['Date of Sale (dd/mm/yyyy)'] = pd.to_datetime(ppr_data_f['Date of Sale (dd/mm/yyyy)'], 
                                   errors = 'coerce')

## Extracting individual features from the field
ppr_data_f['DOS:Year'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.year
ppr_data_f['DOS:Month'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.month
ppr_data_f['DOS:Day'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.day

ppr_data_f['DOS:day_of_year'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.day_of_year
ppr_data_f['DOS:day_of_week'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.day_of_week

ppr_data_f['DOS:is_weekend'] = np.where(ppr_data_f['DOS:day_of_week'].isin([5,6]), 1,0)
ppr_data_f['DOS:is_year_start'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.is_year_start
ppr_data_f['DOS:is_year_end'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.is_year_end
ppr_data_f['DOS:is_quarter_start'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.is_quarter_start
ppr_data_f['DOS:is_quarter_end'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.is_quarter_end
ppr_data_f['DOS:is_month_start'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.is_month_start
ppr_data_f['DOS:is_month_end'] = ppr_data_f['Date of Sale (dd/mm/yyyy)'].dt.is_month_end

## Removing the original date of sale column
del ppr_data_f['Date of Sale (dd/mm/yyyy)']

## Converting the feature column which are numbers to integer type
ppr_data_f["DOS:is_year_start"] = ppr_data_f["DOS:is_year_start"].astype(int)
ppr_data_f["DOS:is_quarter_start"] = ppr_data_f["DOS:is_quarter_start"].astype(int)
ppr_data_f["DOS:is_month_start"] = ppr_data_f["DOS:is_month_start"].astype(int)
ppr_data_f["DOS:is_year_end"] = ppr_data_f["DOS:is_year_end"].astype(int)
ppr_data_f["DOS:is_quarter_end"] = ppr_data_f["DOS:is_quarter_end"].astype(int)
ppr_data_f["DOS:is_month_end"] = ppr_data_f["DOS:is_month_end"].astype(int)
```

#### Cyclic encoding of time-bound features

```python
## Converting the cyclic features to its respective co-sine and sine values
ppr_data_f['DOS:Day' + '_cos'] = np.cos(2 * np.pi * ppr_data_f['DOS:Day']/31)
ppr_data_f['DOS:Day' + '_sin'] = np.sin(2 * np.pi * ppr_data_f['DOS:Day']/31)

ppr_data_f['DOS:Month' + '_cos'] = np.cos(2 * np.pi * ppr_data_f['DOS:Month']/12)
ppr_data_f['DOS:Month' + '_sin'] = np.sin(2 * np.pi * ppr_data_f['DOS:Month']/12)

ppr_data_f['DOS:day_of_week' + '_cos'] = np.cos(2 * np.pi * ppr_data_f['DOS:day_of_week']/6)
ppr_data_f['DOS:day_of_week' + '_sin'] = np.sin(2 * np.pi * ppr_data_f['DOS:day_of_week']/6)

ppr_data_f['DOS:day_of_year' + '_cos'] = np.cos(2 * np.pi * ppr_data_f['DOS:day_of_year']/365)
ppr_data_f['DOS:day_of_year' + '_sin'] = np.sin(2 * np.pi * ppr_data_f['DOS:day_of_year']/365)
```

#### Handling of null values

```python
## Converting the price column to float type from string after removing the comma
ppr_data_f['Price'] = ppr_data_f['Price'].str.replace(',', '').astype(float)

## Manually assigning integer values to each category
ppr_data_f.loc[ppr_data_f['Property Size Description'] == 'greater than or equal to 38 sq metres and less than 125 sq metres', 'Property Size Description'] = 1
ppr_data_f.loc[ppr_data_f['Property Size Description'] == 'greater than 125 sq metres', 'Property Size Description'] = 2
ppr_data_f.loc[ppr_data_f['Property Size Description'] == 'less than 38 sq metres', 'Property Size Description'] = 3
ppr_data_f.loc[ppr_data_f['Property Size Description'] == 'greater than or equal to 125 sq metres', 'Property Size Description'] = 4
ppr_data_f.loc[ppr_data_f['Property Size Description'] == 'greater than 38 square meters and less than 125 square meters', 'Property Size Description'] = 5
ppr_data_f['Property Size Description'] = ppr_data_f['Property Size Description'].astype(float)

## Creating internal train and test data
ppr_null_test = ppr_data_f[ppr_data_f['Property Size Description'].isnull()]
ppr_null_train = ppr_data_f.copy()
ppr_null_train.dropna(inplace=True)
y_ppr_null_train = ppr_null_train['Property Size Description']
x_ppr_null_train = ppr_null_train.drop('Property Size Description', axis=1)

x_ppr_null_test = ppr_null_test.drop('Property Size Description', axis=1)
```

```python
## Applying over sampler technique balance the dataset
ros = RandomOverSampler(random_state=0)
ros.fit(x_ppr_null_train, y_ppr_null_train)
X_resampledros, y_resampledros = ros.fit_resample(x_ppr_null_train, y_ppr_null_train)

## Using random forest model to train on the dataset with no null values
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_resampledros, y_resampledros)
## Predicting values in dataset with null values
pred = clf.predict(x_ppr_null_test)
preddf = pd.DataFrame(pred)
```

```python
# Joining the data with predicted value to the dataset with no null values to form back the original dataframe
x_ppr_null_test = x_ppr_null_test.reset_index(drop=True)
ppr_data_nna = x_ppr_null_test.join(preddf)
ppr_data_nna.rename(columns={0:'Property Size Description'}, inplace=True)

ppr_data_f.dropna(inplace=True)
y_ppr_data_f = ppr_data_f['Property Size Description']
ppr_data_f = ppr_data_f.drop('Property Size Description', axis=1)
ppr_data_f = ppr_data_f.join(y_ppr_data_f)

ppr_data_final = pd.concat([ppr_data_nna, ppr_data_f])
ppr_data_final = ppr_data_final.reset_index(drop=True)

## Converting integer value to its original categorical text state and one-hot encoding the columns
ppr_data_final.loc[ppr_data_final['Property Size Description'] == 1, 'Property Size Description'] = 'greater than or equal to 38 sq metres and less than 125 sq metres'
ppr_data_final.loc[ppr_data_final['Property Size Description'] == 2, 'Property Size Description'] = 'greater than 125 sq metres'
ppr_data_final.loc[ppr_data_final['Property Size Description'] == 3, 'Property Size Description'] = 'less than 38 sq metres'
ppr_data_final.loc[ppr_data_final['Property Size Description'] == 4, 'Property Size Description'] = 'greater than or equal to 125 sq metres'
ppr_data_final.loc[ppr_data_final['Property Size Description'] == 5, 'Property Size Description'] = 'greater than 38 square meters and less than 125 square meters'
encoder_df = pd.DataFrame(encoder.fit_transform(ppr_data_final[['Property Size Description']]).toarray())
encoder_df.columns = encoder.get_feature_names_out(['Property Size Description'])
ppr_data_final = ppr_data_final.join(encoder_df)
ppr_data_final = ppr_data_final.drop('Property Size Description', axis=1)
```

## Data Preparation for data analytics and GIS
In this section the raw data has been manupulated using the steps which were performed during pre-processing and two more datasets were also imported to aim the data analysis.

The address was geo-coded and the final dataframe was saved as a csv file which will be used in Tableau.

```python
## PErforming the pre-processing steps
ppr_data_vis['Description of Property'] = ppr_data_vis['Description of Property'].str.replace('Teach/�ras�n C�naithe Ath�imhe','Restoration Dwelling House /Residence')
ppr_data_vis['Description of Property'] = ppr_data_vis['Description of Property'].str.replace('Teach/�ras�n C�naithe Nua','New Dwelling house /Apartment')
ppr_data_vis.loc[39798,'Description of Property'] = 'New Residence House'
ppr_data_vis.rename(columns={'Price (�)':'Price'}, inplace=True)
ppr_data_vis['Price'] = ppr_data_vis['Price'].str[1:]
ppr_data_vis['Date of Sale (dd/mm/yyyy)'] = pd.to_datetime(ppr_data_vis['Date of Sale (dd/mm/yyyy)'], errors = 'coerce')
ppr_data_vis.rename(columns={'Date of Sale (dd/mm/yyyy)':'Date of Sale'}, inplace=True)
ppr_data_vis['Price'] = ppr_data_vis['Price'].str.replace(',', '').astype(float)
ppr_data_vis = ppr_data_vis[ppr_data_vis['Price']<595000.0]
ppr_data_vis = ppr_data_vis.reset_index(drop=True)
ppr_data_vis = ppr_data_vis[['Date of Sale','Address','County','Price','Description of Property']]
```

```python
ppr_data_vis['Country'] = 'Ireland' ## Adding country column for Tableau to reconise region
ppr_data_vis.to_csv('D:/MTU MSc DSA/Semester 3 - Thesis/data/ppr_data_vis.csv')   ## Saving first dataset for data anlysis
pop_data = pd.read_csv('D:/MTU MSc DSA/Semester 3 - Thesis/data/Pop_Area.csv') ## Reading poulation and county size data
ppr_data_vis_fin = pd.merge(ppr_data_vis, pop_data,  how='left', left_on=['County',ppr_data_vis['Date of Sale'].dt.year], right_on = ['County','Year'])
ppr_data_vis_fin.to_csv('D:/MTU MSc DSA/Semester 3 - Thesis/data/ppr_data_vis.csv')  ## Saving the second data for relationship analysis
```

```python
## Preparing the address column by removing unwanted details in address
ppr_data_vis_fin['min_Add'] = ppr_data_vis_fin['Address'].apply(lambda x: x.strip().split(',')[0])
ppr_data_vis_fin["Ad_Conc"] = ppr_data_vis_fin["min_Add"] + ", " + ppr_data_vis_fin["County"]

## Creating a subset of data with only 9 county's which will be analysed in GIS
ppr_data_vis_selected  = ppr_data_vis_fin[ppr_data_vis_fin['County'].isin(['Mayo','Waterford','Galway','Limerick','Clare','Kerry','Cork','Tipperary'])]
```

```python
## Using Nominatim libraty to geocode the address to latitute and longitude
locator = Nominatim(user_agent="myGeocoder")
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
ppr_data_vis_selected['location'] = ppr_data_vis_selected['Ad_Conc'].apply(geocode)
ppr_data_vis_selected['point'] = ppr_data_vis_selected['location'].apply(lambda loc: tuple(loc.point) if loc else None)
#ppr_data_vis_selected[['latitude', 'longitude', 'altitude']] = pd.DataFrame(ppr_data_vis_selected['point'].tolist(), index=ppr_data_vis_selected.index)
```

```python
## Seperating the latitude, longitude and altitude out of the location points
ppr_data_vis_selected[['latitude', 'longitude', 'altitude']] = pd.DataFrame(ppr_data_vis_selected['point'].apply(pd.Series))
```

```python
## Getting only the data where nominatim found the location for and discarding the remaining
ppr_data_vis_gis = ppr_data_vis_selected.copy()
ppr_data_vis_gisf = ppr_data_vis_gis[ppr_data_vis_gis['latitude'].notna()]

# Saving the final file for GIS analysis in Tableau
ppr_data_vis_gisf.to_csv('D:/MTU MSc DSA/Semester 3 - Thesis/data/ppr_data_vis_gisf.csv')
```
